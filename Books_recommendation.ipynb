## Use Cosine as a metric

```python
book_pivot_table = pd.pivot_table(data = book_data,values='Book_Rating',index='User_ID',columns='Book_Title').fillna(0)
book_pivot_table.index = book_data['User_ID'].unique()
```

```python
book_pivot_table
```

```python
book_pivot_table.values
```

```python
from sklearn.metrics import pairwise_distances
```

```python
cosine_metric = 1 - pairwise_distances(X = book_pivot_table.values,metric='cosine')
cosine_metric
```

```python
cosine_metric         = pd.DataFrame(data=cosine_metric)
cosine_metric.index   = book_data['User_ID'].unique()
cosine_metric.columns = book_data['User_ID'].unique()
```

```python
cosine_metric.round(2)
```

## Filter the data top 50 users

```python
first_50_users = cosine_metric.iloc[:50,:50]
first_50_users
```

```python
first_50_users = np.fill_diagonal(a=first_50_users.to_numpy(),val=0)
first_50_users
```
